```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep, strftime
from random import randint
import pandas as pd
```

```python
chromedriver_path = '/home/rb/Documents/Python Projects/Chrome_Driver/chromedriver_linux64/chromedriver' # Change this to your own chromedriver path!
driver = webdriver.Chrome(executable_path=chromedriver_path)
sleep(2)


driver.get('https://www.greatamericanstations.com/station-listing/')
sleep(3)


#reference the full set of hyperlinks in one and iterate over the list and create link list

link_sections = driver.find_elements_by_xpath("//div[@class='container-wrap']//a[@href]")

links = []

for link in link_sections:
    print(link.get_attribute('href'))
    links.append(link.get_attribute('href'))

station_data =[]

for link in links:
        driver.get(link)
        
        station_name = driver.find_element_by_class_name('station-title').text.split('\n')[0:2]
        ridership = driver.find_element_by_xpath('//*[@id="ajax-content-wrap"]/div[1]/div/div[2]/div[1]/div').text.split('\n')[4:6]

        case = (station_name,ridership)
        station_data.append(case)
        
        sleep(3)
```

```python
import pandas as pd

station_data_df = pd.DataFrame.from_dict(station_data)
```
